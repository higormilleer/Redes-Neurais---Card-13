```python
import pandas as pd
import numpy as np
import tensorflow as tf
import sklearn 
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import utils as np_untils
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
base = pd.read_csv("iris.csv")
```

```python
base
```

```python
x = base.iloc[:,0:4].values
x
```

```python
y = base.iloc[:,4].values
y
```

```python
#transformar as flores em numeros para melhor manipulacao(cada linha um tipo de flor) 
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y
```

```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# Transformar rótulos em números
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)  # Converte rótulos para números inteiros

# Converter para one-hot encoding
y = to_categorical(y)

print("Shape dos rótulos:", y.shape)  # Deve ser (150, 3) para 150 amostras e 3 classes
```

```python
x_treinamento, x_teste, y_treinamento, y_teste =  train_test_split(x,y, test_size = 0.25)
```

```python
x_treinamento.shape, x_teste.shape
```

```python
#3colunas por conta das 3 classes
y_treinamento.shape, y_teste.shape
```

```python
rede_neural = Sequential([
    tf.keras.layers.InputLayer(shape=(4,)),
    tf.keras.layers.Dense(units=4, activation = "relu"), #(4 de entrada + 3 de saida / 2)
    tf.keras.layers.Dense(units=4, activation = "relu"),
    tf.keras.layers.Dense(units=3,activation = "softmax") #3 saidas um para cada iris
])
```

```python
rede_neural.summary()
```

```python
rede_neural.compile(optimizer = "Adam", loss="categorical_crossentropy",metrics=["categorical_accuracy"])
```

```python
rede_neural.fit(x_treinamento, y_treinamento, batch_size=10, epochs=1000)
```

```python
rede_neural.evaluate(x_teste,y_teste)
```

```python
#resultado de todas as probabilidades para cada um dos neuronios, por conta das 3 classes(camada de saida)
previsoes = rede_neural.predict(x_teste)
print(previsoes)
```

```python
previsoes = previsoes > 0.5
print(previsoes)
```

```python
#comparacao do teste com as previsoes
y_teste2 = [np.argmax(t) for t in y_teste]
print(y_teste2)
```

```python
previsoes2 =  [np.argmax(t) for t in previsoes]
print(previsoes2)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_teste2, previsoes2)
```

```python
#setosa, vessicolor, virginica (esta nesta ordem na diagonal principal)
confusion_matrix(y_teste2, previsoes2)
```
