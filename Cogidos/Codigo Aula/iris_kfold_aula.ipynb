```python
import pandas as pd
import tensorflow as tf
import numpy as np
import sklearn 
import scikeras
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
from tensorflow.keras.models import Sequential
from tensorflow.keras import backend as k
from tensorflow.keras import utils as np_utils
```

```python
base = pd.read_csv("iris.csv")
x = base.iloc[:, 0:4].values  
y = base.iloc[:, 4].values  
```

```python
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y)  # Aplica o LabelEncoder em y (1D)
y = np_utils.to_categorical(y)  # Transforma o y codificado em formato categórico
```

```python
def criar_rede():
    k.clear_session()
    rede_neural = Sequential([
    tf.keras.layers.InputLayer(shape=(4,)),  
    tf.keras.layers.Dense(units=4, activation="relu"),  # neurônios na camada oculta
    tf.keras.layers.Dense(units=4, activation="relu"),#mais uma camada oculta
    tf.keras.layers.Dense(units=3, activation="softmax")  # uma saída, portanto sigmoid
])
    rede_neural.compile(optimizer="Adam", loss = "categorical_crossentropy", metrics = ["categorical_accuracy"])
    return rede_neural
```

```python
rede_neural = KerasClassifier(model = criar_rede, epochs=250, batch_size = 10)
```

```python
resultados = cross_val_score(estimator=rede_neural, X=x,y=y, cv=10, scoring="accuracy")
```
