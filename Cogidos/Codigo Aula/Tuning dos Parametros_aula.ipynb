```python
import pandas as pd
import tensorflow as tf
import sklearn 
import scikeras
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras import backend as k
```

```python
X = pd.read_csv("entradas_breast.csv")
y = pd.read_csv("saidas_breast.csv")
```

```python
def criar_rede(optimizer, loss, kernel_initializer, activation, neurons):
    k.clear_session()  # Limpa qualquer sessão anterior
    rede_neural = Sequential([
        tf.keras.layers.InputLayer(input_shape=(30,)),  # Ajuste para o número de características (colunas)
        tf.keras.layers.Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer),
        tf.keras.layers.Dropout(rate=0.2),
        tf.keras.layers.Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer),
        tf.keras.layers.Dropout(rate=0.2),
        tf.keras.layers.Dense(units=1, activation="sigmoid")
    ])
    rede_neural.compile(optimizer=optimizer, loss=loss, metrics=["binary_accuracy"])
    return rede_neural
```

```python
rede_neural_class = KerasClassifier(model=criar_rede)
```

```python
parametros = {
    "model__optimizer": ["Adam", "sgd"],
    "model__loss": ["binary_crossentropy", "hinge"],
    "model__kernel_initializer": ["random_uniform", "normal"],
    "model__activation": ["relu", "tanh"],  # Alterado para 'model__' prefixo
    "model__neurons": [16, 8],
    "batch_size": [10, 30],  
    "epochs": [50, 100]      
}
```

```python
parametros
```

```python
grid_search = GridSearchCV(estimator=rede_neural_class, param_grid=parametros, scoring="accuracy", cv=5)
```

```python
grid_search.fit(X, y)
```
